```python
import matplotlib.pyplot as plt
r0=1e-9
def V(r):
    return cte.e*5*(r/r0)**(1/2)
r=np.linspace(0,100,1000)
plt.axvline(x=0, ymin=0, ymax=3)
plt.title('Potencial')
plt.xlabel('r/r0')
plt.ylabel('V(r)')
plt.plot(r,V(r))
```

```python
def integraldada(x):
    return np.sqrt(1-np.sqrt(x))
resultado= int.quad (integraldada, 0, 1) #igual que lo que dice Mauricio
print('La integral da:', resultado) 
r0=1e-9
V0=cte.e*5/np.sqrt(r0)
CC=0.53333*np.sqrt(2*cte.m_e*V0)*(V0)**(5/2)
E_0=((1/CC*(np.pi*cte.hbar*3/4))**2)**(1/5)
print('Energia BS: ', E_0)
print(CC)
```

```python
radiodecorte=np.sqrt(E/E_0)
print('El radio de corte es: ', radiodecorte)
def inicial(x):
    return np.exp(-x**2/(8*r0**2))/(np.sqrt(8*np.sqrt(np.pi**3)*(r0)**3))
def phi(x):
    return np.sqrt(2/radiodecorte)*np.sin(np.pi*x/radiodecorte)
def c01(x):
    return inicial(x)*phi(x)*x*np.sqrt(4*np.pi)
c0=(int.quad(c01,0,1e-5)[0])**2
print('la probabilidad es: ', c0)


x=np.linspace(0,1e-8,1000)
plt.plot(x,c01(x))
```
